## 1. Conjunto de dados

```python
dominios = ['d_psi', 'd_rs', 'd_amb', 'd_fis']
lbl_dominios = ['Dominio Psicológico', 'Dominio Relações Sociais', 'Dominio Ambiente', 'Dominio Físico']
grupos = ['sexo', 'fxidad']
```

```python
def analise_exploratorio(dataset, dominios, lbl_dominios):
    for i in range(len(dominios)):
        print('\n' + lbl_dominios[i])
        print('--------------------------')
        print(dataset[dominios[i]].describe())
        plt.hist(dataset[dominios[i]], histtype = 'bar', bins = 'auto', color = 'b', 
                 range = (dataset[dominios[i]].min(), dataset[dominios[i]].max()))        
        plt.title('Histograma - ' + lbl_dominios[i])
        plt.xlabel('Métrica ' + lbl_dominios[i])
        plt.ylabel('População')
        plt.show()
```

```python
def super_exploratorio(grupos, vector, dominios, lbl_dominios):
    for i in range(1, len(grupos)):
        print('-------------------------------------------------------------------------')
        print('                          Grupo ' + vector[i+1])
        print('-------------------------------------------------------------------------')
        analise_exploratorio(grupos[i], dominios, lbl_dominios)
```

## 8. Comparativas de Grau de Correlação 

```python
def relatorio(xframe, vector, i):
    plt.matshow(xframe.corr())
    plt.xticks(range(len(xframe.columns)), xframe.columns)
    plt.yticks(range(len(xframe.columns)), xframe.columns)
    plt.title(vector[i] + '\n')
    plt.colorbar()
    plt.show()
```

```python
vector = ['0','Total', 'Homens jovens <16-26]', 'Homens adultos <26-50]', 'Mulheres Jovens <16-26]', 'Mulheres Adultas <26-50]']

def super_relatorio(dominios, vector):
    for i in range(len(vector)):
        if i == 1 :
            xframe = pd.DataFrame(dataset[dominios])
        elif i == 2 :
            xframe = pd.DataFrame(h_jovens[dominios])
        elif i == 3 :
            xframe = pd.DataFrame(h_adultos[dominios])
        elif i == 4 :
            xframe = pd.DataFrame(m_jovens[dominios])
        elif i == 5 :
            xframe = pd.DataFrame(m_adultas[dominios])
        if i != 0:                         
            relatorio(xframe, vector, i)            
```

```python
super_relatorio(dominios, vector)
```
