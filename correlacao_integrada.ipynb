## 1. Conjunto de dados 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fatores = ['idade', 'sexo', 'fxidad', 'd_psi', 'd_rs', 'd_amb', 'd_fis']
dominios = ['d_psi', 'd_rs', 'd_amb', 'd_fis']
grupos = ['sexo', 'fxidad']
```

```python
#dados sem cachorro
dt1 = pd.read_excel('dataset/dados_sc.xls')
el1 = [32, 54]
dt1 = dt1.drop(labels = el1)
dt_sc = pd.DataFrame(dt1, columns = fatores)
print(len(dt_sc) + len(el1))
```

```python
#dados com cachorro
dt2 = pd.read_excel('dataset/dados_cc.xlsx')
el2 = [77, 78, 125]
dt2 = dt2.drop(labels = el2)
dt_cc = pd.DataFrame(dt2, columns = fatores)
print(len(dt_cc) + len(el2))
```

```python
#dados integrados ==> SC : sem cachorro || CC : com cachorro
def rotular(dt, el, tipo):
    if tipo == 0:
        label = 'SC'
    else:
        label = 'CC'
    for i in range(len(dt) + len(el)):
        if i not in el:
            dt['tipo'] = label

rotular(dt_sc, el1, 0)
rotular(dt_cc, el2, 1)
```

```python
dataset = pd.concat([dt_sc, dt_cc], axis = 0, join = 'outer', ignore_index = True, copy = 'true')
dataset
```

## 3. Definição do relatório 

```python
def relatorio(frame, vector, i):
    plt.matshow(frame.corr())
    plt.xticks(range(len(frame.columns)), frame.columns)
    plt.yticks(range(len(frame.columns)), frame.columns)
    plt.title(vector[i] + '\n')
    plt.colorbar()
    plt.show()

#teste
relatorio(pd.DataFrame(dataset[dominios]), vector, 1)
dataset[dominios].corr()
```

```python
#função genérica para geração do relatorio
def super_relatorio(groups, dominios, vector):
    for i in range(len(groups)):        
        relatorio(groups[i][dominios], vector, i+1)        
        print(groups[i][dominios].corr())

#teste
super_relatorio(groups, dominios, vector)
```
