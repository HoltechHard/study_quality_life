## 1. Conjunto de dados

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel('dados_sc.xls')
#limpeza dos dados ==> id = [32, 54] possuem dados faltantes
dataset = dataset.drop(labels = [32, 54], axis = 0)
dataset
```

## 2. Grupo por Sexo

```python
ghomens = dataset[(dataset['sexo'] == 'M')]
gmulheres = dataset[(dataset['sexo'] == 'F')]
total = dataset.groupby(['sexo']).size()
print('H: ', len(ghomens), '\nM: ', len(gmulheres), '\nTotal: ', len(ghomens) + len(gmulheres))
```

## 3. Grupo por Faixa Etária  
Jovens : [16 - 26>  
Adultos: [26 - 50]

```python
jovens = dataset[(dataset['idade']>16) & (dataset['idade']<=26)]
adultos = dataset[(dataset['idade']>26) & (dataset['idade']<=50)]
print('Jovens: ', len(jovens), '\nAdultos: ', len(adultos), '\nTotal: ', len(jovens) + len(adultos))
```

## 4. Histograma de distribuição de população x idade

```python
print('Parametros estatísticos <idade>')
print('-------------------------------')
v_idade = dataset['idade']
v_idade = v_idade.dropna()
v_idade.describe()
```

```python
plt.hist(v_idade, histtype = 'bar',bins = 'auto', color = 'b',range = (v_idade.min(), v_idade.max()))
plt.title('Histograma populacional x idade')
plt.xlabel('Idade')
plt.ylabel('População')
plt.show()
```

## 8. Comparativas de Grau de Correlação 

```python
def relatorio(xframe, vector, i):
    plt.matshow(xframe.corr())
    plt.xticks(range(len(xframe.columns)), xframe.columns)
    plt.yticks(range(len(xframe.columns)), xframe.columns)
    plt.title(vector[i] + '\n')
    plt.colorbar()
    plt.show()
```

```python
vector = ['0','Total', 'Homens jovens [16-26>', 'Homens adultos [26-50]', 'Mulheres Jovens [16-26>', 'Mulheres Adultas [26-50>']

def super_relatorio(dominios, vector):
    for i in range(len(vector)):
        if i == 1 :
            xframe = pd.DataFrame(dataset[dominios])
        elif i == 2 :
            xframe = pd.DataFrame(h_jovens[dominios])
        elif i == 3 :
            xframe = pd.DataFrame(h_adultos[dominios])
        elif i == 4 :
            xframe = pd.DataFrame(m_jovens[dominios])
        elif i == 5 :
            xframe = pd.DataFrame(m_adultas[dominios])
        if i != 0:                         
            relatorio(xframe, vector, i)            
```

```python
super_relatorio(dominios, vector)
```
