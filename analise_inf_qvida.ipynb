## Análise Inferencial ANOVA < análise de variância > 

Definição:  
Trata-se de um método estatístico, desenvolvido por Fisher, que através de testes de igualdade de médias, verifica-se fatores que produzem mudanças sistemáticas em alguma variável de interesse com um certo nivel de significância

Cálculo de parâmetros da ANOVA

```python
param_anova = {}

def somas_amostrais(dataset, fatores):
    somas = {}
    for i in range(3, len(fatores)):
        somas[fatores[i]] = round(dataset[fatores[i]].sum(), 3)
    return somas

param_anova['sum_dt1'] = somas_amostrais(dt1, fatores)
param_anova['sum_dt2'] = somas_amostrais(dt2, fatores)
```

```python
def medias_amostrais(dataset, fatores):
    medias = {}
    for i in range(3, len(fatores)):
        medias[fatores[i]] = round(dataset[fatores[i]].mean(), 3)
    return medias

param_anova['media_dt1'] = medias_amostrais(dt1, fatores)
param_anova['media_dt2'] = medias_amostrais(dt2, fatores)
```

```python
def parametros_total(param_anova, dominios, dt1, dt2):    
    param_anova['sum_total'] = {}
    param_anova['media_total'] = {}
    for i in range(len(dominios)):    
        param_anova['sum_total'][dominios[i]] = round(param_anova['sum_dt1'][dominios[i]] + param_anova['sum_dt2'][dominios[i]], 3)
        param_anova['media_total'][dominios[i]] = round(param_anova['sum_total'][dominios[i]] / (len(dt1) + len(dt2)), 3)

parametros_total(param_anova, dominios, dt1, dt2)        
```

Cálculo de Variâncias Amostrais

```python
def sigma_quadratico(dataset, nome, param_anova, excluidos):    
    param_anova['S2_' + nome] = {}
    for k in range(len(dominios)):
        dataset['s2_' + dominios[k]] = 0.0
    for k in range(len(dominios)):
        for i in range(len(dataset) + len(excluidos)):
            if i not in excluidos:
                dataset['s2_' + dominios[k]][i] = round((dataset[dominios[k]][i] - param_anova['media_' + nome][dominios[k]]) ** 2, 3)
    for k in range(len(dominios)):
        param_anova['S2_' + nome][dominios[k]] = round(dataset['s2_' + dominios[k]].sum() / (len(dataset) - 1), 3)        
```

Cálculo do Se^2

```python
#Entre amostras : calculo Se
def calcular_Se(dt1, dt2, param_anova, dominios):
    param_anova['Qe'] = {}
    param_anova['Se'] = {}
    n1 = len(dt1)
    n2 = len(dt2)
    for k in range(len(dominios)):
        param_anova['Qe'][dominios[k]] = 0.0 
        param_anova['Qe'][dominios[k]] += round(n1 * ((param_anova['media_dt1'][dominios[k]] - param_anova['media_total'][dominios[k]]) ** 2), 3)
        param_anova['Qe'][dominios[k]] += round(n2 * ((param_anova['media_dt2'][dominios[k]] - param_anova['media_total'][dominios[k]]) **2), 3)
        gl = 2 #graus de liberdade sao 2 ==> grupo dt1 || dt2
        param_anova['Se'][dominios[k]] = round((param_anova['Qe'][dominios[k]] / (gl - 1)), 3)

#executar Se ---
calcular_Se(dt1, dt2, param_anova, dominios)
print(param_anova['Se'])
```

Cálculo do Sr^2

```python
#Dentro das amostras : calculo Sr
def calcular_Sr(dt1, dt2, param_anova, dominios):
    param_anova['Qr'] = {}
    param_anova['Sr'] = {}
    gl1 = (len(dt1) - 1)
    gl2 = (len(dt2) - 1)
    for k in range(len(dominios)):
        param_anova['Qr'][dominios[k]] = round(gl1 * param_anova['S2_dt1'][dominios[k]] + gl2 * param_anova['S2_dt2'][dominios[k]], 3)     
        param_anova['Sr'][dominios[k]] = round(param_anova['Qr'][dominios[k]] / (gl1 + gl2), 3)

#executar Sr ---
calcular_Sr(dt1, dt2, param_anova, dominios)
print(param_anova['Sr'])
```

Cálculo do parâmetro F

```python
#Calculo do parametro Fcal
def calcular_F(param_anova, dominios):
    param_anova['Fcal'] = {}
    for k in range(len(dominios)):
        param_anova['Fcal'][dominios[k]] = round((param_anova['Se'][dominios[k]] / param_anova['Sr'][dominios[k]]), 3)

#executar F ---
calcular_F(param_anova, dominios)
print(param_anova['Fcal'])
```

Valor de F a partir da Função de Distribuição de Probabilidade de Fisher 

```python
from IPython.display import Image
Image('img/testeF.jpg')
```

```python
#Calculo do valor F(1, 255) com 5% de significância
F_1_200_95 = 3.89
F_1_400_95 = 3.86
Ftab_1_255_95 =  round((3.89 + (3.86 - 3.89) * ((255 - 200) / (400 - 200))), 3)
print(Ftab_1_255_95)
```

```python
#tabela anova
from astropy.table import Table
import numpy as np

def tabela_anova(param_anova, dominio, nome, gl_qe, gl_qr, Ftab):    
    cabecalho = ['Fonte var', 'Sum. Quadrados', 'Graus liberdade', 'Quadrados medios', 'Teste F']
    tbl_anova = Table(names = cabecalho, meta = {'name' : nome}, dtype = ('S2', 'f8', 'S20', 'S20', 'f8'))
        
    tbl_anova.add_row(('Qe', param_anova['Qe'][dominio], 'K-1 = ' + str(gl_qe), 'Se ^2 = ' + str(param_anova['Se'][dominio]), param_anova['Fcal'][dominio])) 
    tbl_anova.add_row(('Qr', param_anova['Qr'][dominio], 'n - K = ' + str(gl_qr), 'Sr ^2 = ' + str(param_anova['Sr'][dominio]), Ftab))    
    print(tbl_anova)
```

Teste de Hipótese

```python
from IPython.display import Image
Image('img/provaf.jpg')
```

```python
#Teste de Hipótese
def teste_hipotese(param_anova, dominios, gl_qe, gl_qr, Ftab):
    param_anova['msg'] = {}
    for k in range(len(dominios)):
        print('\n Teste de Hipótese: Análise de Variância - Teste F - FATOR: ' + str(dominios[k]) + '\n')
        print('------------------------------------------------------------------ \n')
        if param_anova['Fcal'][dominios[k]] < Ftab:
            msg = 'Aceita-se Ho : u1 = u2, com 5% significância \n Conclusão : Não há diferença entre os grupos no dominio, o fator PRESENÇA DO CACHORRO NÃO INFLUENCIA sobre o dominio ' + dominios[k]            
        else:
            msg = 'Rejeita-se Ho : u1 != u2, com 5% significância \n Conclusão : Há uma diferença significativa entre os grupos no dominio, o fator PRESENÇA DO CACHORRO INFLUENCIA sobre o dominio ' + dominios[k]            
        param_anova['msg'][dominios[k]] = msg
        print(param_anova['msg'][dominios[k]])
        print('\n Tabela ANOVA')
        tabela_anova(param_anova, dominios[k], 'tbl_' + dominios[k], gl_qe, gl_qr, Ftab)    
```

```python
#executar teste hipótese ---
teste_hipotese(param_anova, dominios, 1, 255, Ftab_1_255_95)
```

## Análise ANOVA : diferenciado x grupos 

----------------------------------------------------------------------------------------
Procedimento genérico para execução da prova de hipótese usando ANOVA - distribuição F

```python
def gen_proc_anova_inference(dt1, dt2, grupo1, grupo2, param_anova, ex1, ex2, dominios, fatores, Ftab):
    
    #somatoria dos dados x grupo
    param_anova['sum_dt1'] = somas_amostrais(grupo1, fatores)
    param_anova['sum_dt2'] = somas_amostrais(grupo2, fatores)
    
    #media dos dados x grupo
    param_anova['media_dt1'] = medias_amostrais(grupo1, fatores)
    param_anova['media_dt2'] = medias_amostrais(grupo2, fatores)
    
    #soma e media dos grupos em conjunto
    parametros_total(param_anova, dominios, grupo1, grupo2)
    
    #calculo das variancias <sigma quadrático>
    exc1 = [x for x in dt1.index.values if x not in grupo1.index.values] + ex1
    sigma_quadratico(grupo1, 'dt1', param_anova, exc1)
    exc2 = [x for x in dt2.index.values if x not in grupo2.index.values] + ex2
    sigma_quadratico(grupo2, 'dt2', param_anova, exc2)
    
    #cálculo do Se
    calcular_Se(grupo1, grupo2, param_anova, dominios)
                       
    #calculo do Sr
    calcular_Sr(grupo1, grupo2, param_anova, dominios)
    
    #calculo do parametro F
    calcular_F(param_anova, dominios)
    
    #teste de hipótese F
    teste_hipotese(param_anova, dominios, 1, len(grupo1) + len(grupo2) - 2, Ftab)    

##fim
```
